```python
import sqlite3
conn=sqlite3.connect('my_database.db')
```

```python
df1.to_sql('data1', conn, if_exists='replace', index=False)
```

```python
df2.to_sql('data2', conn, if_exists='replace', index=False)
```

# 1) Display all the details of movies created by directors Christopher and Matt Reeves.

```python
result1=conn.execute("SELECT * FROM data1 WHERE Director1 IN ('Christopher Nolan ', 'Matt Reeves ');")
for i in result1:
    print(i)
```

# 2) Display all the details of movies with a duration of 140 minutes to 190 minutes.

```python
result2=conn.execute("SELECT * FROM data1 WHERE Duration >= 140 AND Duration <= 190;")
for i in result2:
    print(i)
```

# 3) Display all details of movies with ratings above 7 in ascending order.

```python
result3=conn.execute("SELECT * FROM data1 WHERE Rating > 7 ORDER BY Rating ASC;")
for i in result3:
    print(i)
```

# 4) Display all movie names in descending order

```python
result4=conn.execute("SELECT MovieName FROM data1 ORDER BY MovieName DESC; ")
for i in result4:
    print(i)
```

# 5) Display movie name starts with ‘P’ and their rating is greater than 7.

```python
result5=conn.execute("SELECT MovieName FROM data1 WHERE MovieName LIKE 'P%' AND Rating > 7;")
for i in result5:
    print(i)
```
